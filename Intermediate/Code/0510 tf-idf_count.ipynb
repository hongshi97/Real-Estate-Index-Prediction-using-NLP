```python
import pandas as pd

df = pd.read_csv('0429naver_real_estate_noun_verb_adjective.csv', index_col = 'Unnamed: 0')
df
```

```python
spi_df = pd.read_csv('./예측모델/분석용데이터_utf_spi수정완료.csv')
```

```python
df['spi'] = spi_df['spi']
```

```python
df['next_spi'] = df['spi'].shift(-1)
df['next_spi'][-1] = 124.9
```

```python
df['diff_spi'] = df['next_spi']-df['spi']
```

```python
up = df[df['diff_spi']>0]
```

```python
down = df[df['diff_spi']<0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=55601, stop_words='english')
tdm = tfidf.fit_transform(up['hangul'])
up_word_count = pd.DataFrame({
    '단어': tfidf.get_feature_names(),
    '상승tf_idf': tdm.sum(axis=0).flat
})
up_word_count
```

```python
tfidf = TfidfVectorizer(max_features=55601, stop_words='english')
tdm = tfidf.fit_transform(down['hangul'])
down_word_count = pd.DataFrame({
    '단어': tfidf.get_feature_names(),
    '하락tf_idf': tdm.sum(axis=0).flat
})
down_word_count
```

```python
merge_df = pd.merge(up_word_count, down_word_count, on='단어', how='outer')
```

```python
merge_df['dif_idf']=merge_df['상승tf_idf']-merge_df['하락tf_idf']
merge_df
```

```python
merge_df[merge_df['상승tf_idf']>2].sort_values('dif_idf', ascending=False)[:30]
```
